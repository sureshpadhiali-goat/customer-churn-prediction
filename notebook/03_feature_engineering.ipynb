```python
import pandas as pd

df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.to_csv("../data/cleaned_WA_Fn-UseC_-Telco-Customer-Churn.csv", index=False)

df_fe = pd.read_csv("../data/cleaned_WA_Fn-UseC_-Telco-Customer-Churn.csv")
df_fe = df.copy()
df_fe["TenureGroup"] = pd.cut(
    df_fe["tenure"],
    bins=[0, 12, 24, 48, 72],
    labels=["0-1 year", "1-2 years", "2-4 years", "4+ years"]
)
df_fe["HighMonthlyCharge"] = (df_fe["MonthlyCharges"] > df_fe["MonthlyCharges"].median()).astype(int)
df_fe["TotalCharges"] = pd.to_numeric(df_fe["TotalCharges"], errors="coerce")
df_fe["AvgSpendPerMonth"] = df_fe["TotalCharges"] / (df_fe["tenure"] + 1)
df_fe["IsMonthToMonth"] = (df_fe["Contract_Month-to-month"] if "Contract_Month-to-month" in df_fe.columns else 0)
df_fe["IsMonthToMonth"] = df_fe.filter(like="Contract").sum(axis=1)
df_fe["IsElectronicCheck"] = df_fe.filter(like="Electronic check").sum(axis=1)
df_fe["ServiceCount"] = df_fe.filter(like="Service").sum(axis=1)
df_fe["ServiceCount"] = df_fe.filter(regex="Internet|Phone|Streaming").sum(axis=1)
print(df_fe.shape)
df_fe.head()
df_fe = df_fe.dropna(subset=["TotalCharges"])
df_fe["AvgSpendPerMonth"] = df_fe["TotalCharges"] / (df_fe["tenure"] + 1)

```
